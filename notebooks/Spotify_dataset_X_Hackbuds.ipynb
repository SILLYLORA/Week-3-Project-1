```python
import pandas as pd
import numpy as np

df = pd.read_excel(r"C:\Users\chuan\Desktop\Ironhack\week 3\Week 3 Project 1\data\raw\Spotify_Tracks_Dataset.xlsx")

display(df)
```

```python
# Column names 
df.columns

# Convert Column name into title case 
df.columns=df.columns.str.replace('_', ' ').str.title()
#display(df.head())
```

```python
# Check duplicate values 
display(df.duplicated().sum()) #No duplicates

# Check and handle missing values:'Name', 'Artists', 'Album Name', 'Track Name'
df.fillna('Unknown', inplace=True)
df.isnull().sum() 
```

```python
# Filter the df for songs with popularity above 90
filtered_df = df[df['Popularity'] > 90]
sorted_df = filtered_df.sort_values(by='Popularity', ascending=False)

# Create a pivot table to demonstrate filtered songs(Pop.>90)
pivot_table = filtered_df.pivot_table(
    index=['Track Name', 'Artists'],  # Row index
    values='Popularity',    # Values to aggregate
    aggfunc=['count', 'mean'] 
)

pivot_table = pivot_table.sort_values(by=('mean', 'Popularity'), ascending=False)
pivot_table = pivot_table.reset_index()
# Display the pivot table

display(pivot_table)
```

```python

artist_counts = filtered_df['Artists'].value_counts()

most_frequent_artist = artist_counts.idxmax()
most_frequent_artist_count = artist_counts.max()

display(f"The most frequent artist is {most_frequent_artist} with {most_frequent_artist_count} songs.")
```

```python
pivot_table = filtered_df.pivot_table(
    index='Track Name',
    values=['Popularity', 'Duration Ms'],  # Example of including another column
    aggfunc={'Popularity': ['count', 'mean'], 'Duration Ms': 'mean'}
)

pivot_table = pivot_table.reset_index()
print(pivot_table)
```
