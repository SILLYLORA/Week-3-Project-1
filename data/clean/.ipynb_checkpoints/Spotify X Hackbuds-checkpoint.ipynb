```python
import pandas as pd
import numpy as np

df = pd.read_excel(r"C:\Users\chuan\Desktop\Ironhack\week 3\Week 3 Project 1\data\raw\universal_top_spotify_songs.xlsx") 

# Remove the data from 2023 
df['snapshot_date '] = pd.to_datetime(df['snapshot_date'])

df_2024 = df[df['snapshot_date'].dt.year != 2023]

display(df_2024)
```

```python
# Column names 
df_2024.columns

# Convert Column name into title case 
df_2024.columns=df_2024.columns.str.replace('_', ' ').str.title()
display(df_2024.head())
```

```python
# Check duplicate values 
display(df_2024.duplicated().any()) #No duplicates

# Check and handle missing values:'Name', 'Artists', 'Country', 'Album Name', 'Album Release Date'
df_2024.fillna('Unknown', inplace=True)
df_2024.isnull().sum() 
```

```python
# Map countries to regions 
# Define the regions 
Western_Europe=['AT', 'BE', 'CH', 'DE', 'FR', 'GB', 'IE', 'LU', 'NL'],        
Southern_Europe=['ES', 'IT', 'GR', 'PT', 'MT', 'HR', 'SI', 'ME', 'RS', 'BA', 'AL', 'MK', 'BG', 'RO', 'TR', 'CY'],        
Eastern_Europe=['UA', 'SK', 'RS', 'RU', 'PL', 'MD', 'LT', 'LV', 'XK', 'HU', 'GE', 'EE', 'CZ', 'BY', 'AZ', 'AM'],       
Northern_Europe=['DK', 'FI', 'IS', 'NO', 'SE']

def classify_region(Country):
        if Country in Western_Europe:
            return 'Western Europe' 
        elif Country in Southern_Europe:
            return 'Southern Europe'
        elif Country in Eastern_Europe:
            return 'Eastern Europe'
        elif Country in Northern_Europe:
            return 'Northen Europe'
        else:
            return 'Other Regions'

df_2024['Region'] = df_2024['Country'].apply(classify_region)

display(df_2024.head(10))
```
